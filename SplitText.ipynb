### Understanding what we are supposed to do:


*What we have:*

Train dataset has the additional column survived, where 1 means the passenger survived and 0 means the passenger died;
  
*What we should do:*

Test dataset has no such column, which we are expected to **predict** whether those passengers on test dataset **survived or not**.

```python
train.head(3)
```

```python
print(train.isna().sum()) # checking missing values
test.isna().sum()
```

```python
#Cabin has many missing values, let's see;
train['Cabin'].value_counts()
```

```python
# As the existing values also unique, we can drop the Cabin column from both datasets
for df in (train, test): #dropping with less code
    df.drop('Cabin', axis = 1, inplace = True)
```

# Splitting the text and taking necessary info from it

```python
train.head()
```


We have a 'Name' column which is of course different for every single passenger *(meaning unique)*;

Such columns should be dropped because they are not useful;

But we can also see that they have their title in the text, such as Mr, Miss...

So we can first their title, then drop the 'Name' column;

Let's start!


```python
train['title'] = train['Name'].str.split(',', expand = True)[1].str.split('.', expand = True)[0]
```

```python
train['title'].value_counts()
```

```python
train.head()
```

```python
train.drop('Name', axis = 1, inplace = True)
```

```python
train.head()
```

```python
# Let's do the same with test dataframe
test['title'] = test['Name'].str.split(',', expand = True)[1].str.split('.', expand = True)[0]
test.drop('Name', axis = 1, inplace = True)
```
